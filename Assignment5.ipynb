```python
d=dict(zip(f.junior,f.senior))
```

```python
r=[]
for i in d:
    t=[]
    t.append(i)
    tkey=i
    while tkey!=d[tkey]:
        tkey=d[tkey]
        t.append(tkey)
    r.append(t)
    
```

```python
d=p.DataFrame(r,names=(""))
d
```
